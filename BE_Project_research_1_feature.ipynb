```python
import tensorflow as tf
from sklearn.model_selection import train_test_split

# Sample data (replace this with your dataset)
text_data = list(df["SubjectsList"])
labels = list(classes)

# Tokenizing the text data
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(df['SubjectsList'])
sequences = tokenizer.texts_to_sequences(df['SubjectsList'])

# Padding sequences to ensure uniform length
max_length = max([len(seq) for seq in sequences])
padded_sequences = tf.keras.preprocessing.sequence.pad_sequences(sequences, maxlen=max_length, padding='post')

print("Tokenized sequences:")
print(padded_sequences.shape)




```

```python
from sklearn.preprocessing import LabelEncoder
# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the "Domain" column
df['Domain_encoded'] = label_encoder.fit_transform(df['Domain'])

df
```

```python
X=padded_sequences
Y=np.array(df["Domain_encoded"])
print(X.shape)
print(Y.shape)
type(Y)
s=set(Y)
s
```

```python
# Splitting data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
len(X_train)
```

```python
# Model architecture
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(36,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')  # Adjust the output units based on the number of classes in your data
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Training the model
model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test))
```

###KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Initialize the KNN classifier
knn = KNeighborsClassifier(n_neighbors=9)  # You can adjust the number of neighbors (K) here

# Train the KNN classifier
knn.fit(X_train, y_train)

# Predict on the test set
y_pred = knn.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of KNN: {accuracy}")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score

# Example usage after predicting values
precision = precision_score(y_test, y_pred, average='macro')  # Change average for multiclass
recall = recall_score(y_test, y_pred, average='macro')  # Change average for multiclass
f1 = f1_score(y_test, y_pred, average='macro')  # Change average for multiclass
conf_matrix = confusion_matrix(y_test, y_pred)
print("precision : ",precision)
print("recall : ",recall)
print("f1 score : ",f1)
print("conf_matrix : ",conf_matrix)
```
