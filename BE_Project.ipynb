```python
subject_list=[
    "Mathematics"
,"Physics"
,"Chemistry"
,"Biology"
,"Computer Science"
,"Accountancy"
,"Economics"
,"Business Studies"
,"Literature"
,"History"
,"Geography"
,"Sociology"
,"Fine Arts"
,"Political Science"
,"Legal Studies"
,"English"
,"Media Studies"
,"Communication"
,"Music"
,"Dance"
,"Drama/Theatre"
,"Aeronautics"
,"Environmental Science"
,"Sports Science"
,"Psychology"
,"Home Science"
,"Physical Education"
,"Anatomy"
,"Food Science"
,"Library Science"
,"Information Technology"
,"Library Science"
,"Marine Science"
,"Forensic Science"
]
```

```python
subject_dict={}
for subject in subject_list:
  subject_dict[subject]=[None]
subject_dict["Domain"]=[None]
subject_dict["Prefered Domain"]=[None]
```

```python
subject_dict
```

```python
df=pd.DataFrame(subject_dict)
df
```

```python
i=0
for sub,it in zip(data["Subjects of your interest"],data["Current Domain"]):
  s=sub.split(',')

  # print(s)
  record={}
  for item in s:
    its=item.lstrip()
    its=its.rstrip()
    record[item]=int(1)
  record["Domain"]=it
  print(record)
  df.loc[i]=record
  i+=1
```

```python
df_filled = df.fillna(0)
```

```python
df_filled.columns.tolist()
```

```python
df_filled
```

```python
y=df_filled["Domain"]
x=df_filled.drop("Domain",axis=1)
x=x.drop("Prefered Domain",axis=1)
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
